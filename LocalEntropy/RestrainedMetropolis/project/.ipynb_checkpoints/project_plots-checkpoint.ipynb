```python
from time import time
from os import listdir
from os.path import isdir, isfile
import glob
import argparse
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.size': 12})
results_dir = '../results'
processed_dir = '../processed'
plots_dir = '.'
dirlist = [d for d in listdir(f'{results_dir}') if isdir(f'{results_dir}/{d}') and d != '.ipynb_checkpoints' and 'SM' in d]
```

```python
dirlist
```

```python
T_dirlist, gamma_dirlist = [], []
Tlist, gammalist = [], []
for d in dirlist:
    splitted_d = d.split('_')
    T, gamma = float(splitted_d[-2][1:]), float(splitted_d[-1][1:])
    if gamma == 0:
        T_dirlist.append(d)
        Tlist.append(T)
    else:
        gamma_dirlist.append(d)
        gammalist.append(gamma)
        
ord_Tlist = np.sort(Tlist)[::-1]
T_dirlist = [T_dirlist[Tlist.index(T)] for T in ord_Tlist]

ord_gammalist = np.sort(gammalist)
gamma_dirlist = [gamma_dirlist[gammalist.index(T)] for T in ord_gammalist]
```

```python
def load_data(d):
    datafiles = np.array([f for f in listdir(f'{results_dir}/{d}') if isfile(f'{results_dir}/{d}/{f}') and 'data' in f])
    eq_file = datafiles[['eq' in datafile for datafile in datafiles]][0]
    file = datafiles[[not 'eq' in datafile for datafile in datafiles]][0]
    
    with open(f'{results_dir}/{d}/{eq_file}', 'r') as f:
        lines = f.readlines()
    eq_data = np.array([line.split('\t') for line in lines], dtype = float)
    
    with open(f'{results_dir}/{d}/{file}', 'r') as f:
        lines = f.readlines()
    data = np.array([line.split('\t') for line in lines], dtype = float)
    return eq_data, data


def load_muts(d):
    mutsfiles = np.array([f for f in listdir(f'{results_dir}/{d}') if isfile(f'{results_dir}/{d}/{f}') and 'mutants' in f])
    eq_file = mutsfiles[['eq' in mutsfile for mutsfile in mutsfiles]][0]
    file = mutsfiles[[not 'eq' in mutsfile for mutsfile in mutsfiles]][0]
    
    with open(f'{results_dir}/{d}/{eq_file}', 'r') as f:
        lines = f.readlines()
    eq_muts = np.array([[line.split('\t')[0], line.split('\t')[1][:-1]] for line in lines], dtype = str)
    
    with open(f'{results_dir}/{d}/{file}', 'r') as f:
        lines = f.readlines()
    muts = np.array([[line.split('\t')[0], line.split('\t')[1][:-1]] for line in lines], dtype = str)
    
    splitted_d = d.split('_')
    T = float(splitted_d[-2][1:])
    gamma = float(splitted_d[-1][1:])
    return eq_muts, muts, T, gamma
```

### Effective energy

```python
dirlist = T_dirlist
```

```python
partial_dirlist = [
    'SM_1PGB_T0.001_g0.0',
    'SM_1PGB_T0.0005_g0.0',
]
```

```python
fig, ax = plt.subplots(figsize = (15, 7), nrows = 1, ncols = 1)
fig.suptitle('Protein 1PGB Effective energy\n' + r'($\gamma$ = 0.0)', y = 0.96)

for d in partial_dirlist:
    eq_data, data = load_data(d)
    eq_gens, gens = eq_data[:, 0], data[:, 0]
    eq_energy, energy = eq_data[:, 1], data[:, 1]
    T = data[0, -3]
    
    ax.plot(eq_gens, eq_energy, color = 'blue', alpha = 0.3)
    ax.plot(gens, energy, alpha = 0.7, label = f'T = {T}')
    
ax.grid()
ax.legend()
ax.set_xlabel('Generation')
ax.set_ylabel('Effective energy')

plt.savefig(f'{plots_dir}/effective_energy.png', bbox_inches='tight');
```

---
### PAM1 distance

```python
fig, ax = plt.subplots(figsize = (15, 7), nrows = 1, ncols = 1)
fig.suptitle('Protein 1PGB Scaled PAM1 distance\n' + r'($\gamma$ = 0.0)', y = 0.96)

for d in partial_dirlist:
    eq_data, data = load_data(d)
    eq_gens, gens = eq_data[:, 0], data[:, 0]
    eq_distance, distance = eq_data[:, 3], data[:, 3]
    T, length = data[0, -3], data[0, -1]
    
    
    ax.plot(eq_gens, eq_distance/length, color = 'blue', alpha = 0.3)
    ax.plot(gens, distance/length, alpha = 0.7, label = f'T = {T}')
    
ax.grid()
ax.legend()
ax.set_xlabel('Generation')
ax.set_ylabel('Scaled PAM1 distance')

plt.savefig(f'{plots_dir}/PAM1_distance.png', bbox_inches='tight');
```

```python
def same_plot(Tlist, energy_threshold = 1., dtype = 'Hamm', prec = 1., savefig = True):
    fig, ax = plt.subplots(figsize = (15, 8), nrows = 1, ncols = 1)
    
    if dtype == 'Hamm':
        fig.suptitle(f'Protein 1PGB group Hamming distances distributions', y = 0.96)
        
        for T in Tlist:
            filenames = [f for f in listdir(f'{processed_dir}/hole_search/pdfs/')
                         if isfile(f'{processed_dir}/hole_search/pdfs/{f}') and (str(T) in f)]
            if len(filenames) == 1:
                filename = filenames[0]
                step = int(filename.split('_')[-1][1:-4])
            elif len(filenames) > 1:
                steps = [int(f.split('_')[-1][1:-4]) for f in filenames]
                filename = filenames[np.argmin(steps)]
                step = np.min(steps)
            else:
                raise ValueError(f'There is no pdf file for {T} temperature.')
            
            pdf = pd.read_csv(f'{processed_dir}/hole_search/pdfs/{filename}')
            distances = pdf['distances'] / pdf['distances'].max()
            pdf = pdf.drop(columns = ['Unnamed: 0', 'distances']) 

            if len(pdf.columns) > 1:
                for column in pdf.columns:
                    ax.plot(distances, pdf[column], label = f'T = {T}, step = {step}, {column}')
            else:
                for column in pdf.columns:
                    ax.plot(distances, pdf[column], label = f'T = {T}, step = {step}')

        ax.set_xlabel('Scaled Hamming distance')
        ax.set_ylabel('pdf')
        ax.legend()
        ax.grid()

    if savefig: plt.savefig(f'{plots_dir}/same_{dtype}_pdfs.png', bbox_inches='tight');
```

```python
Tlist = [
    0.001,
    0.0005
]
energy_threshold = 1.
dtype = 'Hamm'
prec = 1.
```

```python
same_plot(Tlist, energy_threshold, dtype, prec)
```
