```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import argparse
from os import listdir
from os.path import isfile
```

```python
discarded_mutations = 1000
results_dir = 'RestrainedMetropolis/results/highgamma'
plots_dir = 'plots/highgamma'

plt.rcParams.update({'font.size': 13})
```

```python
def sort_filelist(filelist):
    gammas = []
    for filename in filelist:
        splitted_filename = filename.split('/')[-1]
        gamma_str = splitted_filename.split('_')[-1]
        gamma = float( gamma_str[1] + '.' + gamma_str[2:-4])
        gammas.append(gamma)
    ordered_gammas = np.sort(gammas)
    filelist = [filelist[gammas.index(gamma)] for gamma in ordered_gammas][::-1]
    return filelist

data_files = [filename for filename in listdir(f'{results_dir}') if isfile(f'{results_dir}/{filename}') and ('data' in filename)]
data_files = sort_filelist(filelist = data_files)
```

## Equilibrium check

```python
def load_data(filename):
    with open(f'{results_dir}/{filename}', 'r') as file:
        lines = file.readlines()
    splitted_lines = np.array([line.split('\t') for line in lines]).astype(float)
    return splitted_lines
```

```python
def hist_equilibrium(filename, discarded_mutations, savefig_bool = True):
    file_data = load_data(filename)
    T_str = str(file_data[0, -3])[:1] + str(file_data[0, -3])[2:7]
    t_str = str(file_data[0, -4])[:1] + str(file_data[0, -4])[2:7]
    g_str = str(file_data[0, -2])[:1] + str(file_data[0, -2])[2:7]
    file_id = 'T' + T_str + '_t' + t_str + '_g' + g_str
    
    eff_energy = file_data[discarded_mutations:, 1]
    PAM1_dist = file_data[discarded_mutations:, 3]
    mutations = len(eff_energy)
    sections = ['first', 'second', 'last']

    fig = plt.figure(figsize = (12, 8))
    fig.suptitle(f'Effective energy (top) and PAM1 distance (bottom) distribution through the simulation \n(T = {format(file_data[0, -3], ".3f")} - gamma = {format(file_data[0, -2], ".1f")} - threshold = {int(file_data[0, -4])})', y = 0.96)

    ax = plt.subplot(2, 1, 1)
    for idx, section in enumerate(sections):
        section_eff_energy = eff_energy[int(idx * mutations/3):int((idx+1) * mutations/3)]
        ax.hist(section_eff_energy, bins = np.linspace(0., np.max(section_eff_energy), 20), alpha = 0.5, density = True, label=f'{section} third')
    ax.set_ylabel('distribution')
    ax.set_xlabel(r'E$_{eff}$')
    ax.legend(loc = 'best')
    ax.grid(True)
    
    ax = plt.subplot(2, 1, 2)
    for idx, section in enumerate(sections):
        section_PAM1_dist = PAM1_dist[int(idx * mutations/3):int((idx+1) * mutations/3)]
        ax.hist(section_PAM1_dist, bins = np.linspace(0., np.max(section_PAM1_dist), 20), alpha = 0.5, density = True, label=f'{section} third')
    ax.set_ylabel('distribution')
    ax.set_xlabel(r'd$_{PAM1}$')
    ax.legend(loc = 'best')
    ax.grid(True)

    if savefig_bool: plt.savefig(f'{plots_dir}/hist_equilibrium{file_id}.png', bbox_inches='tight')
        
        
def hist_all(data_files, discarded_mutations, savefig_bool = True):
    for filename in data_files:
        hist_equilibrium(filename, discarded_mutations, savefig_bool)
```

```python
hist_all(data_files, discarded_mutations)
```

## PAM1 (and Hamming) distance check

```python
def load_lists(data_files, discarded_mutations, return_Hamming = True):
    gammas, PAM1dlists = [], []
    if return_Hamming: Hammdlists = []
    
    for data_file in data_files:
        with open(f'{results_dir}/{data_file}', 'r') as file:
            lines = file.readlines()
        gamma = float(lines[0].split('\t')[-2])
        PAM1dlist = [float(line.split('\t')[3]) for line in lines]
        if return_Hamming: Hammdlist = [float(line.split('\t')[4]) for line in lines]

        gammas.append(gamma)
        PAM1dlists.append(PAM1dlist[discarded_mutations:])
        if return_Hamming: Hammdlists.append(Hammdlist[discarded_mutations:])

    if return_Hamming:
        return gammas, PAM1dlists, Hammdlists
    else:
        return gammas, PAM1dlists
```

```python
gammas, PAM1dlists, Hammdlists = load_lists(data_files, discarded_mutations)
```

```python
fig = plt.figure(figsize = (12, 8))
plt.title(r'Frequency of the explored distances for different $\gamma$')

for PAM1dlist, gamma in zip(PAM1dlists[::-1], gammas[::-1]):
    PAM1dexpl, counts = np.unique(PAM1dlist, return_counts = True)
    plt.scatter(PAM1dexpl, counts/counts.sum(), label = f'gamma {gamma}', s = 60)

plt.xlabel(r'$d_{PAM1}$')
plt.ylabel('frequency')
plt.legend()
plt.grid(True)

plt.savefig(f'{plots_dir}/wt_frequency.png', bbox_inches='tight');
```

```python
def reweighted_mean(dlist, old_gamma, new_gamma):
    dlist = np.array(dlist)
    denominator = np.exp(-(new_gamma - old_gamma) * dlist)
    numerator = dlist * denominator
    return np.mean(numerator) / np.mean(denominator)

def explored_means(dlists, gammas, return_matrix = False):
    means_matrix = np.zeros((len(gammas), len(gammas)), dtype = float)
    for row in range(len(gammas)):
        means_matrix[row, row] = np.mean(dlists[row])

        for col in range(len(gammas)):
            if col == row: 
                continue
            means_matrix[row, col] = reweighted_mean(dlists[col], gammas[col], gammas[row])

    means = np.mean(means_matrix, axis = 1)
    if return_matrix:
        return means, means_matrix
    else:
        return means

def predicted_means(dlists, gammas, new_gammas):
    means = []
    for new_gamma in new_gammas:
        mean = 0.
        for dlist, old_gamma in zip(dlists, gammas):
            mean += reweighted_mean(dlist, old_gamma, new_gamma)
        means.append(mean / len(gammas))
    return np.array(means)
```

```python
new_gammas = np.linspace(0., 9, 201)
mask = [not gamma in gammas for gamma in new_gammas]
new_gammas = new_gammas[mask]

explored_PAM1_means, means_matrix = explored_means(PAM1dlists, gammas, return_matrix = True)
predicted_PAM1_means = predicted_means(PAM1dlists, gammas, new_gammas)
ss_PAM1_means = np.array([means_matrix[row, row] for row in range(len(means_matrix))])
explored_Hamm_means = explored_means(Hammdlists, gammas, return_matrix = False)
```

```python
fig = plt.figure(figsize = (12, 8))
ax = plt.subplot(1, 1, 1)
ax.set_title(r'$\overline{d_{PAM1}}$  and  $\overline{d_{Hamm}}$  (with $d_{PAM1}$ based simulations)')

ax.scatter(gammas, explored_PAM1_means, color = 'red', label = 'PAM1 simulated d (with reweighting)', marker = 'o', s = 100)
ax.scatter(gammas, ss_PAM1_means, color = 'orange', label = 'PAM1 simulated d (without reweighting)', marker = 's', s = 60)
ax.scatter(gammas, explored_Hamm_means, color = 'blue', label = 'Hamm simulated d (with reweighting)', marker = '^', s = 60)
ax.scatter(new_gammas, predicted_PAM1_means, color = 'green', label = 'PAM1 predicted d (with reweighting)', marker = 'o', s = 15)
ax.plot([new_gammas.min(), new_gammas.max()], [0., 0.], linestyle = '--', color = 'grey')

ax.set_xlabel(r'$\gamma$')
ax.set_ylabel(r'$\overline{d}$')
ax.grid(True)
ax.legend(handlelength = 5)

plt.savefig(f'{plots_dir}/reweighting.png', bbox_inches='tight');
```
