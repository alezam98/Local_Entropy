```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DistPAM1.csv')
df = df.drop(columns = ['Unnamed: 0'])
df
```

```python
distances = np.array(df).reshape(-1)
distances = distances[distances != 0]
```

```python
plt.rcParams.update({'font.size': 13})
bins = np.linspace(0., 1, 21)
formattedbins = [format(bin, '.2f') for bin in bins]

fig = plt.figure(figsize = (12, 8))
ax = plt.subplot(1, 1, 1)
ax.set_title('PAM1 distance distribution')
ax.hist(distances, bins = bins, density = True)
ax.set_xlabel(r'd$_{PAM1}$')
ax.set_ylabel(r'$\rho$')
ax.set_xticks(bins)
ax.set_xticklabels(formattedbins, rotation = '70');
```
